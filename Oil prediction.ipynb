```python
my_dict = {'Rio Grande do Norte':52809601,
          'Sergipe':2068017,
          'Bahia':565733,
          'Rio de Janeiro':7780.8,
          'Amazonas':3483985,
          'Alagoas':27845.1,
          'Espírito Santo':46077.5,
          'Ceará':146348.3,
          'Paraná':199298,
          'Maranhão':329651.496,
          'São Paulo':53369.61}
```

```python
df['state_weights'] = df['State'].map(my_dict)
```

```python
df['oil_per_hr_for_state'] = ((df['Óleo (bbl/dia)'])/(df['Production time (hrs per month)']))/(df['state_weights']/1000)
```

```python
df['oil_per_state'] = (df['Óleo (bbl/dia)'])/(df['state_weights']/1000)
df
```

```python
df = df.fillna(0)
```

```python
plt.plot(df['Period'],df['oil_per_hr_for_state'])
```

```python
plt.plot(df['Period'],df['oil_per_state'])
```

```python
train = df.iloc[:1400000]
test = df.iloc[1400000:]
```

```python

import pandas as pd
from prophet import Prophet

# Prepare the data in the required format for Prophet
data = train.rename(columns={'Period': 'ds', 'oil_per_state': 'y'})

# Create and fit the Prophet model
model = Prophet()
model.fit(data)

# Generate future dates for prediction
future_dates = model.make_future_dataframe(periods=365)  # Specify the number of future periods

# Make predictions for both training and future dates
predictions = model.predict(future_dates)

# Extract the predicted values for the training set
predicted_training = predictions[:len(data)]

# Visualize the predictions for training and future dates
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['ds'], data['y'], label='Actual')
plt.plot(predicted_training['ds'], predicted_training['yhat'], label='Predicted (Training)')
plt.plot(predictions['ds'], predictions['yhat'], label='Predicted (Future)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Prophet Forecast')
plt.legend()
plt.show()
```
